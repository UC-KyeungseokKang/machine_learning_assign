```python
import matplotlib.pyplot as plt
import numpy as np

file_data		= "mnist_train.csv"
handle_file	= open(file_data, "r")
data        		= handle_file.readlines()
handle_file.close()

size_row	= 28    # height of the image
size_col  	= 28    # width of the image

num_image	= len(data)
count       	= 0     # count for the number of images


# normalize the values of the input data to be [0, 1]

def normalize(data):

    data_normalized = (data - min(data)) / (max(data) - min(data))

    return(data_normalized)
```

```python
list_image  = np.zeros((((size_row * size_col)+1), num_image), dtype=float)
list_label  = np.zeros(num_image, dtype=int)


label_check0 = np.zeros(num_image,dtype=int)
label_check1 = np.zeros(num_image,dtype=int)
label_check2 = np.zeros(num_image,dtype=int)
label_check3 = np.zeros(num_image,dtype=int)
label_check4 = np.zeros(num_image,dtype=int)
label_check5 = np.zeros(num_image,dtype=int)
label_check6 = np.zeros(num_image,dtype=int)
label_check7 = np.zeros(num_image,dtype=int)
label_check8 = np.zeros(num_image,dtype=int)
label_check9 = np.zeros(num_image,dtype=int)


conclusion0 = np.zeros((size_row * size_col)+1, dtype=float)
conclusion1 = np.zeros((size_row * size_col)+1, dtype=float)
conclusion2 = np.zeros((size_row * size_col)+1, dtype=float)
conclusion3 = np.zeros((size_row * size_col)+1, dtype=float)
conclusion4 = np.zeros((size_row * size_col)+1, dtype=float)
conclusion5 = np.zeros((size_row * size_col)+1, dtype=float)
conclusion6 = np.zeros((size_row * size_col)+1, dtype=float)
conclusion7 = np.zeros((size_row * size_col)+1, dtype=float)
conclusion8 = np.zeros((size_row * size_col)+1, dtype=float)
conclusion9 = np.zeros((size_row * size_col)+1, dtype=float)


for line in data:

    line_data   = line.split(',')
    label       = line_data[0]
    im_vector   = np.zeros((size_row * size_col)+1, dtype=float)
    im_vector[1:]   = np.asfarray(line_data[1:])
    im_vector   = normalize(im_vector)
    im_vector[0] = 1

    list_label[count]       = label
    list_image[:,count]    = im_vector
    #label chcking
    if int(label) == 0:
        label_check0[count] = 1
    elif int(label) == 1:
        label_check0[count] = -1
        label_check1[count] = 1
        label_check2[count] = -1
        label_check3[count] = -1
        label_check4[count] = -1
        label_check5[count] = -1
        label_check6[count] = -1
        label_check7[count] = -1
        label_check8[count] = -1
        label_check9[count] = -1
    elif int(label) == 2:
        label_check0[count] = -1
        label_check1[count] = -1
        label_check2[count] = 1
        label_check3[count] = -1
        label_check4[count] = -1
        label_check5[count] = -1
        label_check6[count] = -1
        label_check7[count] = -1
        label_check8[count] = -1
        label_check9[count] = -1
    elif int(label) == 3:
        label_check0[count] = -1
        label_check1[count] = -1
        label_check2[count] = -1
        label_check3[count] = 1
        label_check4[count] = -1
        label_check5[count] = -1
        label_check6[count] = -1
        label_check7[count] = -1
        label_check8[count] = -1
        label_check9[count] = -1
    elif int(label) == 4:
        label_check0[count] = -1
        label_check1[count] = -1
        label_check2[count] = -1
        label_check3[count] = -1
        label_check4[count] = 1
        label_check5[count] = -1
        label_check6[count] = -1
        label_check7[count] = -1
        label_check8[count] = -1
        label_check9[count] = -1
    elif int(label) == 5:
        label_check0[count] = -1
        label_check1[count] = -1
        label_check2[count] = -1
        label_check3[count] = -1
        label_check4[count] = -1
        label_check5[count] = 1
        label_check6[count] = -1
        label_check7[count] = -1
        label_check8[count] = -1
        label_check9[count] = -1
    elif int(label) == 6:
        label_check0[count] = -1
        label_check1[count] = -1
        label_check2[count] = -1
        label_check3[count] = -1
        label_check4[count] = -1
        label_check5[count] = -1
        label_check6[count] = 1
        label_check7[count] = -1
        label_check8[count] = -1
        label_check9[count] = -1
    elif int(label) == 7:
        label_check0[count] = -1
        label_check1[count] = -1
        label_check2[count] = -1
        label_check3[count] = -1
        label_check4[count] = -1
        label_check5[count] = -1
        label_check6[count] = -1
        label_check7[count] = 1
        label_check8[count] = -1
        label_check9[count] = -1
    elif int(label) == 8:
        label_check0[count] = -1
        label_check1[count] = -1
        label_check2[count] = -1
        label_check3[count] = -1
        label_check4[count] = -1
        label_check5[count] = -1
        label_check6[count] = -1
        label_check7[count] = -1
        label_check8[count] = 1
        label_check9[count] = -1
    elif int(label) == 9:
        label_check0[count] = -1
        label_check1[count] = -1
        label_check2[count] = -1
        label_check3[count] = -1
        label_check4[count] = -1
        label_check5[count] = -1
        label_check6[count] = -1
        label_check7[count] = -1
        label_check8[count] = -1
        label_check9[count] = 1

    count += 1
#transpose for dot process
transpose = np.zeros((count,(size_row * size_col)+1), dtype=float)
transpose = np.transpose(list_image)
#make conclusion 10 times
conclusion0 = np.dot(np.linalg.pinv(transpose), label_check0)
conclusion1 = np.dot(np.linalg.pinv(transpose), label_check1)
conclusion2 = np.dot(np.linalg.pinv(transpose), label_check2)
conclusion3 = np.dot(np.linalg.pinv(transpose), label_check3)
conclusion4 = np.dot(np.linalg.pinv(transpose), label_check4)
conclusion5 = np.dot(np.linalg.pinv(transpose), label_check5)
conclusion6 = np.dot(np.linalg.pinv(transpose), label_check6)
conclusion7 = np.dot(np.linalg.pinv(transpose), label_check7)
conclusion8 = np.dot(np.linalg.pinv(transpose), label_check8)
conclusion9 = np.dot(np.linalg.pinv(transpose), label_check9)
```

```python
true=0;
false=0;
fx_list = np.zeros(10, dtype=float)
# make list and check with np.argmax
for i in range(num_image):
    fx_list[0]=np.dot(transpose[i],conclusion0)
    fx_list[1]=np.dot(transpose[i],conclusion1)
    fx_list[2]=np.dot(transpose[i],conclusion2)
    fx_list[3]=np.dot(transpose[i],conclusion3)
    fx_list[4]=np.dot(transpose[i],conclusion4)
    fx_list[5]=np.dot(transpose[i],conclusion5)
    fx_list[6]=np.dot(transpose[i],conclusion6)
    fx_list[7]=np.dot(transpose[i],conclusion7)
    fx_list[8]=np.dot(transpose[i],conclusion8)
    fx_list[9]=np.dot(transpose[i],conclusion9)
    if np.argmax(fx_list) == list_label[i]:
        true+=1
    else:
        false+=1
#2. Compute (1) true positive rate, (2) error rate using (2) training dataset.
print("true " +str(true))
print("false "+ str(false))
print("true positive rate : {:.1f}% ".format(true/num_image*100))
print("Error rate : {:.1f}% ".format(false/num_image*100))
```
