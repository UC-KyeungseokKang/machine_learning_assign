```python
#library import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read data by pandas
data = pd.read_csv("C:/Users/rkdru/Desktop/test/mnist_train.csv")
```

```python
#The first matrix of the dataset is also a value, but it is treated as a label and divided into x and y.
#The number of the first column in the first column is 5, so we assign the first column to y, then drop it and put the remaining value in X.
y=data['5']
X=data.drop(columns=['5'])
X=np.array(X)
y=np.array(y)
```

```python
#Make a figure for 10 pictures to see the pictures continuously.
fig=plt.figure(figsize=(10, 10))

#Define result_sum to receive result
result_sum=np.zeros(X.shape[1]).astype(float)

#  double for statement about i,j , and calculates the value for each.
for j in range(0,10):
    result_sum=np.zeros(X.shape[1]).astype(float)
    for i in np.where(y==j)[0]:
        result_sum=result_sum+X[i]**2
    result_sum=result_sum**(1/2)
    result_sum=result_sum.reshape(28,28)
    fig.add_subplot(2,5,j+1)
    plt.imshow(result_sum,cmap='Greys', interpolation='None')
plt.show()
```
