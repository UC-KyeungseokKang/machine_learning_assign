# Import packages for project

```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.linalg as lin

plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')
```

# Define function for generating coordinates

```python
num     = 500
std     = 5
result  = 0

def fun(x):
    f =  np.cos(x)*x 
    return f
```

# Define function for getting approximation

```python
def approx(x,y,p):
    arr = np.ones((x.shape[0],p))
    for i in range(p):
        arr.T[i] = x**i;
    result = np.dot(np.dot(lin.inv(np.dot(arr.T,arr)),arr.T),y)
    return np.dot(arr,result)
```

# generate noisy&clean y-coordinate data

```python
n       = np.random.rand(num)
nn      = n - np.mean(n)
x       = np.linspace(-10,10,num)
y1      = fun(x)+nn*std #(noisy) y-coordinate data
y2      = fun(x) # (clean) y-coordinate data
```

```python
plt.plot(x,y1,'k.')
plt.title('A Line without noise')
plt.show()
```

```python
plt.plot(x,y2,'k.')
plt.title('A Line without noise')
plt.show()
```

```python
fig = plt.figure()
fig1 = plt.subplot(111)

curves = approx(x, y1, 0)
string = "p="+str(0)
plt.plot(x,y1,'b.',x,y2,'k',x,curves,'c')
fig1.plot(x, curves, label=string)

box = fig1.get_position()
fig1.set_position([box.x0, box.y0, box.width * 0.8, box.height])
fig1.legend(loc='center left', bbox_to_anchor=(1, 0.5))
fig1.set_title('solve the least square approximate(p=0)')
plt.show()
```
